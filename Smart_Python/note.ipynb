**a mod b‚ü∫ a ‚àí ( a ‚à£ b )√ó b**

```python
# ÂÆö‰∫é‰∏Ä‰∏™ÂáΩÊï∞ËøòÂéüÊï¥Èô§ÁÆóÊ≥ï
def mod(a,b):
    return a - (a // b)*b

# È™åËØÅ‰∏§‰∏™ÁÆóÊ≥ïÊòØÂê¶Á≠â‰ª∑
print(41%14 == mod(41,14))
print(14%41 == mod(14,41))
print(-32%9 == mod(-32,9))
print(32%-9 == mod(32,-9))
```

### Áü≠Ë∑ØÊ±ÇÂÄº Short-Circuit Evaluation

#### ÈÄªËæëËøêÁÆóÂèÇÁÖßË°®:
| X | Y | X and Y | X or Y | not X | not Y |
| --- | --- | --- | --- | --- | --- |
| True | True | True | True | False | False |
| True | False | False | True | False | True |
| False | False | False | False | True | True |
| False | True | False | True | True | False |

üí°Âú®andËøêÁÆó‰∏≠Ôºå Âè™Ë¶ÅXÂá∫Áé∞‰∫ÜFalseÔºåÂ∞±‰∏ÄÂÆöÊòØFalseÔºå‰∏ç‰ºöÁªßÁª≠Âà§Êñ≠Y„ÄÇ

```python
def yes():
    return True

def no():
    return False

def crash():
    return 1/0 # ‰ºöÂ¥©Ê∫ÉÔºÅ
```

```python
print(no() and crash()) # ÊàêÂäüËøêË°å --->False | andËøêÁÆó‰∏≠,Êúâ‰∏Ä‰∏™ÊòØFalse,ÁªìÊûúÂ∞±‰∏∫false
```

```python
print(crash() and no()) # Â¥©Ê∫É 
print(yes() and crash()) # Áî±‰∫é‰∏ä‰∏ÄË°åÂ¥©Ê∫É,ÂàôËøô‰∏ÄË°å‰∏ç‰ºöË¢´ËøêË°å
```

üí°Âú®andËøêÁÆó‰∏≠Ôºå Âè™Ë¶ÅXÂá∫Áé∞‰∫ÜTrueÔºåÂ∞±‰∏ÄÂÆöÊòØTrueÔºå‰∏ç‰ºöÁªßÁª≠Âà§Êñ≠Y„ÄÇ

```python
print(yes() or crash()) # ÊàêÂäüËøêË°å ---> True | orËøêÁÆó‰∏≠,Êúâ‰∏Ä‰∏™ÊòØTrue,ÁªìÊûúÂ∞±ÊòØTrue
```

```python
print(crash() or yes()) # Â¥©Ê∫É 
print(no() or crash())  # Áî±‰∫é‰∏ä‰∏ÄË°åÂ¥©Ê∫É,ÂàôËøô‰∏ÄË°å‰∏ç‰ºöË¢´ËøêË°å
```

```python
def isPositive(n):
    result = (n > 0)
    print(n, "ÊòØ‰∏çÊòØÊ≠£Êï∞?", result)
    return result
def isEven(n):
    result = (n%2==0)
    print(n, "ÊòØ‰∏çÊòØÂÅ∂Êï∞?", result)
    return result
```

```python
print(isEven(-4) and isPositive(-4)) #Ë∞ÉÁî®‰∫Ü‰∏§‰∏™ÂáΩÊï∞
```

```python
print(isEven(-3) and isPositive(-3)) #Âè™Ë∞ÉÁî®‰∫Ü‰∏Ä‰∏™ÂáΩÊï∞
```

```python
print(isEven(-4) or isPositive(-4)) # Âè™Ë∞ÉÁî®‰∫Ü‰∏Ä‰∏™ÂáΩÊï∞
```

```python
def isNumber(x):
    return ((type(x) == int) or
            (type(x) == float))
```

```python
print(isNumber(1), isNumber(1.1), isNumber(1+2j), isNumber("p2s")) 
## typeÊòØÂà§Êñ≠‰∏ç‰∫ÜÂ§çÊï∞ÁöÑ
```
